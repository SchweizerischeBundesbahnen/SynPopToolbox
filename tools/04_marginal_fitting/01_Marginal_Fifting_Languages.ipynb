## Fixing methodology

### Expected Counts

```python
expected_pop_counts = (marginals['expected_counts_{}'.format(YEAR)].rename('expected_counts')
                       .reset_index()
                       .pivot(index='N_Gem', columns='language', values='expected_counts')
                       )
```

### Fixing ...

The logging can be set to `INFO` or `DEBUG` (at the top of the notebook `logger.setLevel(logging.INFO)`) based on how much detail is needed!

**This is going to take a few minutes to run!**

```python
fixed_language = marginal_fitting.fix_categorical_feature(persons=persons,
                                                          feature='language',
                                                          pop_segment_variables=['N_Gem'],
                                                          control_totals=expected_pop_counts,
                                                          person_proba=None  # totally random
                                                          )
```

```python
persons_fixed = persons.copy(deep=True)
persons_fixed = persons_fixed.drop('language', axis=1)
persons_fixed = pd.merge(persons_fixed, fixed_language, left_on='person_id', right_index=True)
```

## Examining SynPop+

### Global Counts

```python
%%time
summary_table_fixed = marginal_fitting.compute_comparison_summary(persons_fixed, persons_ist, 
                                                                  YEAR, YEAR_IST, 
                                                                  groupby='language')
```

```python
print('As reference, the total population growth is: {:.2f}'.format(persons.shape[0] / persons_ist.shape[0]))
summary_table_fixed
```

### Marginals per Commune

```python
%%time 
marginals_fixed = marginal_fitting.compute_ist_vs_scenario_marginal_summary_table(persons_fixed, persons_ist,
                                                                                  YEAR, YEAR_IST, 
                                                                                  feature='language', 
                                                                                  control_level_list=['N_Gem'])
```
