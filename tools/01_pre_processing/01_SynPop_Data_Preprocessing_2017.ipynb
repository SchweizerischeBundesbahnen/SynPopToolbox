## Global Settings

```python
DATA_DIR = r'\\k13536\mobi\10_Daten\SynPop'
RAW_SYNPOP_DATA_DIR = os.path.join(DATA_DIR, '2017/190813_CH_2017_BEST1/01_raw')
PROCESSED_DATA_DIR = os.path.join(DATA_DIR, '2017/190813_CH_2017_BEST1/02_preprocessed')

MOBI_ZONES_SHP_PATH = os.path.join(DATA_DIR, '_mobi_zones/mobi_zones.shp')

BLUR_HOUSEHOLDS = True

YEAR = 2017
```

## Running Preprocessor

```python
preprocessor = SynPopPreprocessor(RAW_SYNPOP_DATA_DIR, PROCESSED_DATA_DIR, YEAR, MOBI_ZONES_SHP_PATH)
```

### Persons & Households

```python
persons, households = preprocessor.preprocess_persons_and_households(blur_households=BLUR_HOUSEHOLDS)
```
